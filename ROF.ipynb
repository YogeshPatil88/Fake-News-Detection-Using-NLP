```python
import pandas as pd
from sklearn import metrics
import numpy as np
import itertools
from sklearn.metrics import confusion_matrix
```

```python
df=pd.read_csv('fake_or_real_news.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df.label.replace({"FAKE":0 ,"REAL":0 }, inplace = True)
```

```python
df.head()
```

```python
x=df.drop('label',axis=1)
x.head()
```

```python
y=df['label']
y.head()
```

```python
df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
messages=df.copy()
```

```python
messages['title'][6]
```

```python
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
corpus[3]
```

# CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,ngram_range=(1,3))  #take the most  5000 frequently occuring words and use 1-3 words to make a feature
x = cv.fit_transform(corpus).toarray()
```

```python
x.shape
```

```python
y=messages['label']
```

```python
y.shape
```

```python
x
```

```python
y
```

```python
## Divide the dataset into Train and Test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=88)
```

```python
cv.get_feature_names()
```

```python
cv.get_params()
```

```python
count_df = pd.DataFrame(x_train, columns=cv.get_feature_names())
count_df.head(20)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

### Naive Bayes

```python
from sklearn.naive_bayes import MultinomialNB
nbc=MultinomialNB()
nbc.fit(x_train, y_train)
pred = nbc.predict(x_test)
tab1=confusion_matrix(pred,y_test)
tab1
```

```python
acc=tab1.diagonal().sum()* 100 / tab1.sum()
acc
```

```python
feature_names = cv.get_feature_names()
sorted(zip(nbc.coef_[0], feature_names))[:]
```

# TF-IDF

```python
## TFidf Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(max_features=5000,ngram_range=(1,3)) ###
x=tfidf.fit_transform(corpus).toarray()
x
```

```python
## Divide the dataset into Train and Test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
count_tfidf = pd.DataFrame(x_train, columns=tfidf.get_feature_names())
count_tfidf.head(20)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import MultinomialNB
nbc=MultinomialNB()
nbc.fit(x_train, y_train)
pred2 = nbc.predict(x_test)
tab2=confusion_matrix(pred2,y_test)
tab2
```

```python
acc2=tab2.diagonal().sum()* 100 / tab2.sum()
acc2
```
